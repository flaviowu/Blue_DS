```python
import pandas as pd
import plotly.express as px
import math
```

```python
musicas = pd.read_csv("Musicas.csv")
```

## Funçao que calcula a distância euclidiana dados uma referência

```python
def get_dist(e1, e2, d1, d2):
    dist = math.sqrt(((e1-e2)**2) + ((d1-d2)**2))
    return dist
```

```python
s = musicas.sample(3)[["danceability", "energy"]]  #Sample 3 musicas
s.reset_index(drop=True, inplace=True)
s
```

```python
musicas.head()
```

```python
musicas = musicas.loc[:,["song_title", "artist", "danceability", "energy"]]
musicas.head()
```

```python
s.at[1, "danceability"]
```

```python
for index, row in musicas.iterrows():
    musicas.at[index, "dist_0"] = get_dist(row["danceability"], s.at[0, "danceability"], row["energy"], s.at[0, "energy"])
    musicas.at[index, "dist_1"] = get_dist(row["danceability"], s.at[1, "danceability"], row["energy"], s.at[1, "energy"])
    musicas.at[index, "dist_2"] = get_dist(row["danceability"], s.at[2, "danceability"], row["energy"], s.at[2, "energy"])
musicas.head()
```

```python
def get_closest(row):
    if row["dist_0"] < row["dist_1"] and row["dist_0"] < row["dist_2"]:
        return 0
    elif row["dist_1"] < row["dist_0"] and row["dist_1"] < row["dist_2"]:
        return 1
    else:
        return 2
```

```python
for index, row in musicas.iterrows():
    musicas.at[index, "closest"] = get_closest(row)
musicas
```

```python
fig = px.scatter(musicas, x='energy', y='danceability', color='closest')
fig.show()
```
