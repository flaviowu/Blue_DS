```python
import pandas as pd
```

```python
df = pd.read_csv('dados_tratados.csv')
df.info()
```

```python
df['time'] = df['time'].str.replace('não', 'sem time')
```

```python
df['time'] = df['time'].fillna('N/R')
df
```

```python
df['ovo'] = df['ovo'].fillna(df['ovo'].mode()[0])
df['catchup_pizza'] = df['catchup_pizza'].str.upper()
```

```python
df = df.drop(columns = ['nova_coluna'])
df
```

```python
df['time']
```

```python
def mapeamento(x):
    if(x == 'VAI CURINTIA'):
        return 'curintia'
    else:
        return x
```

```python
def mapeamento2(a):
    if(a == 'curintia'):
        return 'VAI CURINTIA'
    else:
        return a
```

```python
df['time2'] = df['time'].map(mapeamento)
df
```

```python
df['time3'] = df['time2'].apply(mapeamento2)
```

```python
df['data_nasc'] = pd.to_datetime(df['data_nasc'])
df['day'] = df['data_nasc'].dt.day
df['month'] = df['data_nasc'].dt.month
df['year'] = df['data_nasc'].dt.year
```

```python
df
```

```python
def get_uf(x):
    uf = x.split('/')[-1]
    return uf
```

```python
df["uf"] = df["localidade"].apply(get_uf)
df
```

```python
df = df.drop(['uf', 'time2', 'time3'], axis=1)
```

```python
def get_sign(x):
    if x["month"] == 12:
        astro_sign = 'Sagitário' if (x['day'] < 22) else 'Carpicórnio'
    elif x["month"] == 1:
        astro_sign = 'Carpicórnio' if (x['day'] < 20) else 'Aquário'
    elif x["month"] == 2:
        astro_sign = 'Aquário' if (x['day'] < 19) else 'Peixes'
    elif x["month"] == 3:
        astro_sign = 'Peixes' if (x['day'] < 21) else 'Áries'
    elif x["month"] == 4:
        astro_sign = 'Áries' if (x['day'] < 20) else 'Touro'
    elif x["month"] == 5:
        astro_sign = 'Touro' if (x['day'] < 21) else 'Gêmeos'
    elif x["month"] == 6:
        astro_sign = 'Gêmeos' if (x['day'] < 21) else 'Câncer'
    elif x["month"] == 7:
        astro_sign = 'Câncer' if (x['day'] < 23) else 'Leão'
    elif x["month"] == 8:
        astro_sign = 'Leão' if (x['day'] < 23) else 'Virgem'
    elif x["month"] == 9:
        astro_sign = 'Virgem' if (x['day'] < 23) else 'Libra'
    elif x["month"] == 10:
        astro_sign = 'Libra' if (x['day'] < 23) else 'Escorpião'
    elif x["month"] == 11:
        astro_sign = 'Escorpião' if (x['day'] < 22) else 'Sagitário'
    return astro_sign
```

```python
df['signo'] = df.apply(get_sign, axis=1)
```

```python
df.at[19,'localidade'] = 'Paulínia/SP'
df['uf'] = df['localidade'].apply(get_uf)
df
```
