```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel('./Pumpkin_Seeds_Dataset.xlsx')
df.head()
```

```python
df = pd.get_dummies(data=df, columns= ['Class'], drop_first=True)
```

```python
df.head()
```

```python
df.corr()[['Class_Ürgüp Sivrisi']]
```

```python
#sns.pairplot(data=df, hue='Class')
```

```python
from sklearn.preprocessing import StandardScaler

from sklearn.model_selection import train_test_split
from sklearn.feature_selection import SelectKBest, f_classif
from sklearn.linear_model import LogisticRegression
from sklearn.naive_bayes import GaussianNB

from sklearn.metrics import confusion_matrix, accuracy_score, precision_score, recall_score
```

```python
X = df.drop(['Class_Ürgüp Sivrisi'], axis=1)
y = df['Class_Ürgüp Sivrisi']
```

```python
anova =  SelectKBest(f_classif, k = 5).fit(X, y)
X_new = df[anova.get_feature_names_out()]
X_std = StandardScaler().fit_transform(X_new)
y = df['Class_Ürgüp Sivrisi']
anova.get_feature_names_out() # visualiza as variáveis mais importantes
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_std, y, test_size = 0.2, random_state=10)
```

## Naive Bayes

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred), 
             columns=["Predito Ürgüp Sivrisi","Predito não Ürgüp Sivrisi",],
             index  =["Verdadeiro Ürgüp Sivrisi", "Verdadeiro não Ürgüp Sivrisi"])
```

```python
print(round(accuracy_score(y_test, y_pred), 4))
print(round(precision_score(y_test, y_pred, pos_label=1), 4))
print(round(recall_score(y_test, y_pred, pos_label=1), 4))
```

```python
metrics = {"nb_std": { 
        "acuracia": accuracy_score(y_test, y_pred),
        "precisao": precision_score(y_test, y_pred, pos_label=1),
        "recall": recall_score(y_test, y_pred, pos_label=1)
        }
    }
```

## Regressão logística

```python
lr = LogisticRegression(random_state=10)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred), 
             columns=["Predito Ürgüp Sivrisi","Predito não Ürgüp Sivrisi",],
             index  =["Verdadeiro Ürgüp Sivrisi", "Verdadeiro não Ürgüp Sivrisi"])
```

```python
print(round(accuracy_score(y_test, y_pred), 4))
print(round(precision_score(y_test, y_pred, pos_label=1), 4))
print(round(recall_score(y_test, y_pred, pos_label=1), 4))
```

```python
metrics.update({"lr_std": { 
        "acuracia": accuracy_score(y_test, y_pred),
        "precisao": precision_score(y_test, y_pred, pos_label=1),
        "recall": recall_score(y_test, y_pred, pos_label=1)
        }
    })
```

## PCA

```python
from sklearn.decomposition import PCA
```

```python
pca = PCA(n_components=12)
X_pca = pca.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_pca, y, test_size = 0.2, random_state=10)
```

```python
gnb = GaussianNB()
gnb.fit(X_train, y_train)
y_pred = gnb.predict(X_test)
```

```python
pd.DataFrame(confusion_matrix(y_test, y_pred), 
             columns=["Predito Ürgüp Sivrisi","Predito não Ürgüp Sivrisi",],
             index  =["Verdadeiro Ürgüp Sivrisi", "Verdadeiro não Ürgüp Sivrisi"])
```

```python
print(round(accuracy_score(y_test, y_pred), 4))
print(round(precision_score(y_test, y_pred, pos_label=1), 4))
print(round(recall_score(y_test, y_pred, pos_label=1), 4))
```

```python
metrics.update({"nb_pca": { 
        "acuracia": accuracy_score(y_test, y_pred),
        "precisao": precision_score(y_test, y_pred, pos_label=1),
        "recall": recall_score(y_test, y_pred, pos_label=1)
        }
    })
```

```python
lr = LogisticRegression(random_state=10)
lr.fit(X_train, y_train)
y_pred = lr.predict(X_test)
```

```python
print(round(accuracy_score(y_test, y_pred), 4))
print(round(precision_score(y_test, y_pred, pos_label=1), 4))
print(round(recall_score(y_test, y_pred, pos_label=1), 4))
```

```python
metrics.update({"lr_pca": { 
        "acuracia": accuracy_score(y_test, y_pred),
        "precisao": precision_score(y_test, y_pred, pos_label=1),
        "recall": recall_score(y_test, y_pred, pos_label=1)
        }
    })
```

```python
metrics_df = pd.DataFrame(metrics).T.reset_index()
```

```python
metrics_df.rename({ 'index': 'modelo'}, axis=1, inplace=True)
```
